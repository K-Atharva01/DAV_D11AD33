```python
#Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
data = pd.read_csv('Dataset/city_day.csv')

# Display the first few rows of the dataset to understand its structure
print(data.head())
```

```python
data.describe()
```

```python
#Percetage of missing values
missing_percent = 100 * data.isnull().sum() / len(data)
missing_percent
```

```python
data.interpolate(limit_direction="both",inplace=True)
```

```python
#Percetage of missing values
missing_percent = 100 * data.isnull().sum() / len(data)
missing_percent
```

```python
for i,each in enumerate(data['AQI_Bucket']):
    if pd.isnull(data['AQI_Bucket'].iloc[i]):
        if data['AQI'].iloc[i]>=0.0 and data['AQI'].iloc[i]<=50.0:
            data['AQI_Bucket'].iloc[i]='Good'
        elif data['AQI'].iloc[i]>=51.0 and data['AQI'].iloc[i]<=100.0:
            data['AQI_Bucket'].iloc[i]='Satisfactory'
        elif data['AQI'].iloc[i]>=101.0 and data['AQI'].iloc[i]<=200.0:
            data['AQI_Bucket'].iloc[i]='Moderate'
        elif data['AQI'].iloc[i]>=201.0 and data['AQI'].iloc[i]<=300.0:
            data['AQI_Bucket'][i]='Poor'
        elif data['AQI'].iloc[i]>=301.0 and data['AQI'].iloc[i]<=400.0:
            data['AQI_Bucket'].iloc[i]='Very Poor'
        else:
            data['AQI_Bucket'].iloc[i]='Severe'
```

```python
data.isnull().sum()
```

```python
# Plotting Subset of Cities
cities_to_plot = ['Mumbai', 'Chennai', 'Delhi']  # Example: Top 5 cities by population

plt.figure(figsize=(12, 8))
sns.lineplot(x='Date', y='AQI', data=data[data['City'].isin(cities_to_plot)], hue='City', palette='Set2', ci=None)
plt.title('AQI Trends Over Time for Selected Cities')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.xticks(rotation=90)
plt.legend(title='City', loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='AQI', data=data, hue='City')
plt.title('AQI Trends Over Time for Each City')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.xticks(rotation=45)
plt.legend(title='City', loc='upper left')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['AQI'], bins=30, kde=True, color='blue')
plt.title('Distribution of AQI Values')
plt.xlabel('AQI')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='AQI_Bucket', y='AQI', data=data, palette='viridis')
plt.title('Distribution of AQI Values Across AQI Buckets')
plt.xlabel('AQI Bucket')
plt.ylabel('AQI')
plt.show()
```

```python
pollutants = data[['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']]
correlation_matrix = pollutants.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Pollutants')
plt.show()
```

```python
average_aqi = data.groupby('City')['AQI'].mean().sort_values(ascending=False)
plt.figure(figsize=(10, 6))
average_aqi.plot(kind='bar', color='skyblue')
plt.title('Average AQI Values for Each City')
plt.xlabel('City')
plt.ylabel('Average AQI')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PM2.5', y='AQI', data=data)
plt.title('Relationship between AQI and PM2.5 Levels')
plt.xlabel('PM2.5')
plt.ylabel('AQI')
plt.show()
```

```python
pollutant_columns = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']
data['Total_Pollutants'] = data[pollutant_columns].sum(axis=1)
data['PM_Contribution'] = data['PM2.5'] / data['Total_Pollutants']
data['NOx_Contribution'] = data['NOx'] / data['Total_Pollutants']
pollutant_contributions = data[['PM_Contribution', 'NOx_Contribution']]
plt.figure(figsize=(10, 6))
pollutant_contributions.plot(kind='bar', stacked=True)
plt.title('Contribution of Pollutants to AQI')
plt.xlabel('Sample')
plt.ylabel('Contribution')
plt.show()
```

```python
# Calculate the contribution of each pollutant to AQI
pollutant_columns = ['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3', 'Benzene', 'Toluene', 'Xylene']
data['Total_Pollutants'] = data[pollutant_columns].sum(axis=1)
for col in pollutant_columns:
    data[col + '_Contribution'] = data[col] / data['Total_Pollutants']

# Plot the contribution of pollutants to AQI
plt.figure(figsize=(12, 8))
pollutant_contributions = data[[col + '_Contribution' for col in pollutant_columns]]
pollutant_contributions.plot(kind='bar', stacked=True, figsize=(12, 8), colormap='viridis')
plt.title('Contribution of Pollutants to AQI')
plt.xlabel('Sample')
plt.ylabel('Contribution')
plt.legend(title='Pollutant')
plt.show()
```

```python
delhi_data = data[data['City'] == 'Delhi']
aqi_bucket_counts = delhi_data['AQI_Bucket'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(aqi_bucket_counts, labels=aqi_bucket_counts.index, autopct='%1.1f%%', colors=sns.color_palette('Pastel1', len(aqi_bucket_counts)))
plt.title('Distribution of AQI Buckets in Delhi')
plt.show()
```

```python
# Filter data for Delhi
delhi_data = data[data['City'] == 'Delhi']

# Convert 'Date' column to datetime format and set it as index
delhi_data['Date'] = pd.to_datetime(delhi_data['Date'])
delhi_data.set_index('Date', inplace=True)

# Resample data to monthly frequency and calculate mean AQI for each month
delhi_monthly_aqi = delhi_data['AQI'].resample('M').mean()

# Plot the time series of mean monthly AQI for Delhi
plt.figure(figsize=(12, 6))
delhi_monthly_aqi.plot(marker='o', linestyle='-')
plt.title('Mean Monthly AQI in Delhi')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.grid(True)
plt.show()
```
