```r
import matplotlib.pyplot as plt

age=[25,50,75]
income=[25,100,30]
plt.plot(age,income)
plt.title("Age vs Income Salary", fontsize=25, color="green")
plt.ylabel("Income per month (In Thousands)")
plt.xlabel("Ages")
plt.show()
```

# **R:-**

ggplot2 Library

```r
install.packages("ggplot2")
```

```r
library(ggplot2)
```

```r
# Sample data
x <- c(1, 2, 3, 4, 5)
y <- c(2, 4, 5, 4, 3)

# Basic scatter plot
ggplot(data.frame(x, y), aes(x = x, y = y)) + geom_point()
```

```r
# Sample data
x <- c(1, 2, 3, 4, 5)
y <- c(1, 3, 5, 4, 2)

# Line plot
ggplot(data.frame(x, y), aes(x = x, y = y)) + geom_line()
```
